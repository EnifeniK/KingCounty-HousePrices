```python
# Feature Engineering
```

```python
dff = dff.drop('id', axis=1)
```

```python
dff
```

```python
dff['date'] = pd.to_datetime(dff['date'])
```

```python
dff['month'] = dff['date'].apply(lambda date: date.month)
dff['year'] = dff['date'].apply(lambda date: date.year)
```

```python
dff = dff.drop('date', axis=1)
```

```python
dff.head()
```

```python
dff = dff.drop('zipcode', axis=1)
```

```python
dff.head(5)
```

```python
dff['sqft_basement'].value_counts()
```

```python
dfy = dff
```

```python
basement = pd.get_dummies(dff['sqft_basement'])
```

```python
def class_basement(col):
        if col == 0:
            return 0
        else:
            return 1
```

```python
dff['sqft_basement'] = dff['sqft_basement'].apply(class_basement)
```

```python
dff.head()
```

```python
dff['basement'] = dff['sqft_basement']
```

```python
dff = dff.drop('sqft_basement', axis=1)
```

```python
dff.head(5)
```

```python
dff['yr_renovated'].value_counts()
```

```python
def reno(col):
    if col <= 2000:
        return 0
    else:
        return 1
```

```python
dff['renovated'] = dff['yr_renovated'].apply(reno)
```

```python
dff = dff.drop('yr_renovated', axis=1)
```

```python
dff
```

```python
# Training the model
```

```python
X = dff.drop('price', axis=1).values
```

```python
y = dff['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
# Creating the model
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
# Fitting model to train and test data
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), batch_size=128, epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
mean_squared_error(y_test, predictions)
```

```python
mean_absolute_error(y_test, predictions)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
explained_variance_score(y_test, predictions)
```
